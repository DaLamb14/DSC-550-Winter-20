##### 1. Create an array of size [10,50] and use the random function to fill it with integers between 1 and 500 (inclusive). Load it to a DataFrame.

```python
import numpy as np
import pandas as pd
```

```python
array1 = np.random.randint(1,501, size = (10,50))
array1
```

```python
df1 = pd.DataFrame(array1, index = ["Row 1","Row 2","Row 3","Row 4","Row 5","Row 6","Row 7","Row 8","Row 9","Row 10"], columns = list(range(1,51)))
df1.head()
```

```python
df1.columns = df1.columns.astype(str)
df1 = df1.rename(columns=lambda x: 'Col ' + x) 
```

##### 3. Determine the minimum; maximum; average; and standard deviation of all entries.

```python
array2 = array1.flatten()
```

```python
df2 = pd.DataFrame(array2)
print("Minimum:", df2.min())
print("Maximum:", df2.max())
print("Average:", df2.mean())
print("Standard Dev:", df2.std())
```
