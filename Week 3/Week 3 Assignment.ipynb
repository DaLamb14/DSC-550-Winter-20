#### Upload Data

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
iris = load_iris()
df = pd.DataFrame(iris["data"], columns = iris["feature_names"])
df["target"] = iris["target"]
df.head()
```

#### Figure One: Histogram Subplot

```python
fig, axs = plt.subplots(2, 2, tight_layout = True)
axs[0,0].hist(df["sepal length (cm)"], bins = 12, ec = "black", color = "red")
axs[0,0].set_title("Histogram of Sepal Length")
axs[0,0].set_xlabel("Sepal Length (cm)")
axs[0,0].set_ylabel("Frequency")
axs[0,0].set_ylim([0,40])

axs[0,1].hist(df["sepal width (cm)"], bins = 12, ec = "black", color = "green")
axs[0,1].set_title("Histogram of Sepal Width")
axs[0,1].set_xlabel("Sepal Width (cm)")
axs[0,1].set_ylabel("Frequency")
axs[0,1].set_ylim([0,40])

axs[1,0].hist(df["petal length (cm)"], bins = 12, ec = "black", color = "red")
axs[1,0].set_title("Histogram of Petal Length")
axs[1,0].set_xlabel("Petal Length (cm)")
axs[1,0].set_ylabel("Frequency")
axs[1,0].set_ylim([0,40])

axs[1,1].hist(df["petal width (cm)"], bins = 12, ec = "black", color = "green")
axs[1,1].set_title("Histogram of Petal Width")
axs[1,1].set_xlabel("Petal Width (cm)")
axs[1,1].set_ylabel("Frequency")
axs[1,1].set_ylim([0,40])
```

#### Figure Two: Box and Whisker Subplot

```python
data = [df["sepal length (cm)"], df["sepal width (cm)"], df["petal length (cm)"], df["petal width (cm)"]]
plt.boxplot(data,vert = 0)
plt.title("Box and Whisker of Iris Characteristics")
plt.yticks([1,2,3,4], ["Sepal Length (cm)", "Sepal Width (cm)", "Petal Length (cm)", "Petal Width (cm)"])
plt.ylabel("Characteristic")
plt.xlabel("(cm)")
```

#### Figure Three: Density Plots

```python
ax = sns.kdeplot(df["sepal length (cm)"], shade = True, color = "red")
ax.set(xlabel = "Centimeters", ylabel = "Frequency", title = "Density Plot Sepal Length")
```

```python
ax = sns.kdeplot(df["sepal width (cm)"], shade = True, color = "green")
ax.set(xlabel = "Centimeters", ylabel = "Frequency", title = "Density Plot Sepal Width")
```

```python
ax = sns.kdeplot(df["petal length (cm)"], shade = True, color = "red")
ax.set(xlabel = "Centimeters", ylabel = "Frequency", title = "Density Plot Petal Length")
```

```python
ax = sns.kdeplot(df["petal width (cm)"], shade = True, color = "green")
ax.set(xlabel = "Centimeters", ylabel = "Frequency", title = "Density Plot Petal Width")
```

#### Summary

   I feel the most effective figure in this set is the combination box and whisker plot. This figure compares the distribution of data between the different iris characteristics. By judging where the box is comparative to the others, you can understand the size of sepals and petals. The widest distribution of sizes is the petal length, the smallest or most condensed distribution is the sepal width. Sepal width also has the potential for outliers, dots beyond the whiskers, which would have to be checked before further analysis. 
   
   The second most effective figure is the histograms, showing the frequency of values in each bin of measure. In this figure, the sepal length and width seem to have an approximately normal distribution with one point of high frequency and low ends on each side. Unlike the sepal, the petals appear more bimodal with two high points with a  valley in the middle. Analyses done with the petal measurements will have to be checked for the assumption of normal distribution. Further tests can affirm whether or not non-parametric analyses will have to be done, or if a transformation of the data will normalize the data. 
    
   The least effective in the set, but still effective is the Kernel Density Plots. It shows the distribution similarly to the histograms, but smoother. It reflects a similar story to the histograms, the sepals having more unimodal, normal distribution while the petals have less normal, bimodal distributions. Because of the smoothness of these graphs, it is harder to tell the frequency at each centimeter. Having the bins of the histogram gives a little more detailed picture of the distribution.
