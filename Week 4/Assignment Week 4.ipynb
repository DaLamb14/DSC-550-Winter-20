#### Sentiment Analysis

I will be using the TextBlob sentiment analysis for this assignment. The sentiment analysis of the TextBlob function displays two analyses of the statements. The first is polarity, meaning the positivity or negativity of the statement. This measurement ranges from -1, which is very negative, to 1, which is very positive. This means a neutral statement lies at 0. The second measurement displayed is the subjectivity of the statement, meaning the persons emotion, judgement, or opinion in the sentence. This measurement is from 0 to 1 with 1 being the highest amount of subjectivity and 0 being no subjectivity. 

```python
def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text
```

```python
DC["nopunc"] = DC['comments'].apply(remove_punctuations)
DC["clean"] = DC["nopunc"].str.lower()
DC
```

```python
no_punc = DC["nopunc"]
clean_com = DC["clean"]
```

```python
for comments in comment:
    print(comments)
    analysis = TextBlob(comments)
    print(analysis.sentiment)
```

#### AirBnb Reviews

```python
reviews = pd.read_csv(r"C:\Users\datre\OneDrive\Documents\Graduate School\Winter '20\reviews.csv")
reviews.head()
```

```python
del reviews["listing_id"]
del reviews["id"]
del reviews["date"]
del reviews["reviewer_id"]
```

```python
rev_samp = reviews.sample(n=100)
rev_samp.head()
```

```python
pd.set_option("display.max_rows", None)
```

```python
pol = lambda x: TextBlob(x).sentiment.polarity
sub = lambda x: TextBlob(x).sentiment.subjectivity

rev_samp["polarity"] = rev_samp["comments"].apply(pol)
rev_samp["subjectivity"] = rev_samp["comments"].apply(sub)
rev_samp
```
